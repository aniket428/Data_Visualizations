Understanding Attribute Types Using Python

```python
# ï‚· Represent the following:
# o Nominal attributes (e.g., names, departments)
# o Binary attributes (e.g., gender, pass/fail)
# o Ordinal attributes (e.g., low < medium < high)
# o Numeric attributes (continuous and discrete) 

# Q.4 Understanding Attribute Types Using Python
# Representing Nominal, Binary, Ordinal, and Numeric attributes

import pandas as pd
from pandas.api.types import CategoricalDtype


df = pd.read_csv("attribute_example.csv")

print("Dataset:\n", df, "\n")

# -----------------------------
# Step 2: Nominal Attributes
# -----------------------------
print("Nominal Attributes (Name, Department):")
print(df[["Name", "Department"]].head(), "\n")

# -----------------------------
# Step 3: Binary Attributes
# -----------------------------
df["Gender_binary"] = df["Gender"].map({"Male": 1, "Female": 0})
df["Pass_binary"] = df["Pass/Fail"].map({"Pass": 1, "Fail": 0})
print("Binary Attributes (Gender, Pass/Fail):")
print(df[["Gender", "Gender_binary", "Pass/Fail", "Pass_binary"]], "\n")

# # -----------------------------
# # Step 4: Ordinal Attributes
# # -----------------------------
performance_order=CategoricalDtype(categories=["Low","Medium","High"],ordered=True)
df['Performance_order']=(df['Performance'].astype(performance_order).cat.codes)
df


# # -----------------------------
# # Step 5: Numeric Attributes
# # -----------------------------
print("Numeric Attributes:")
print("Age (Discrete):", df["Age"].dtype)
print("Marks (Continuous):", df["Marks"].dtype)
```
